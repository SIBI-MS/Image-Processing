```python
import numpy as np
pic=imageio.imread('/content/drive/MyDrive/dip/Images/demo_2.jpg')
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(15,15))
for c,ax in zip(range(3),ax):
  spl_img=np.zeros(pic.shape,dtype='uint8')
  spl_img[:,:,c]=pic[:,:,c]
  ax.imshow(spl_img)
```
